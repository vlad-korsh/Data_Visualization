### Можно сделать только вертикальные или только горизонтальные панели, если использовать конструкцию типа (". ~ <категория>")

### Что лучше использовать: цвета или панели? А для большого набора?

Цвета лучше подходят для небольшого набора, когда всё можно удобно представить на одном графике. Если выборка очень большая, лучше разделить её по панелям.

### Что делает geom_col()? Чем он отличается от geom_bar()?

geom_col() - отображает суммарное значение целевого (y) признака для каждого возможного значения (x) у всех наблюдений 

geom_bar() - отображает количество наблюдений с одним и тем же значением какого-либо признака

```python
# Например, суммарное значение карат для каждой группы
ggplot(data=diamonds) +\
geom_col(mapping=aes(x="cut", y = 'carat'))
```

### Что произойдёт, если

### запустить без "group=1"? Почему?

```python
ggplot(data=diamonds) + geom_bar(mapping=aes(x="cut", y="..prop.."), group=1)
```

```python
ggplot(data=diamonds) + geom_bar(mapping=aes(x="cut", y="..prop.."))
```

### Зачем тут coord_fixed()? Что делает geom_abline()?

coord_fixed() - это задание сторон для системы координат графика
geom_abline() - диагональная линия с заданным наклоном и сдвигом

```python
ggplot(data=mpg, mapping=aes(x="cty", y="hwy")) +\
geom_point() +\
geom_abline() +\
coord_fixed()
```
